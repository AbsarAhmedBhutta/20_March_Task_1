```python
from faker import Faker
import pandas as pd
import numpy as np
import hashlib
from PIL import Image
from PIL import ImageDraw, ImageFont
import hashlib
from PIL import Image, ImageDraw, ImageFont
from IPython.display import Image as Iamg
from IPython.display import display
```

```python
fake = Faker('ar_SA')
```

```python

def generate_signature(Name):
    img = Image.open('black.jpg')
    draw = ImageDraw.Draw(img)
    text = f"{Name}"
    font = ImageFont.truetype("arial.ttf", 14)
    x = (img.width - 90) // 2
    y = (img.height - 50) // 2
    draw.text((x, y), text, font=font, fill=(255, 255, 255))
    return img

data = []
for _ in range(100):
    Name = fake.name()
    FatherName = fake.name()
    ID = fake.iana_id()
    Address = fake.address()
    Signature = generate_signature(Name)
    Signature.save(f"{Name}.jpg")  # save signature image
    data.append([Name, FatherName, ID, Address, f"{Name}.jpg"])  # append file path to data

    
    
df = pd.DataFrame(data, columns=['Name', 'FatherName', 'ID', 'Address', 'Signature'])
df

```
